**open jupyter from bat file**

to open jupyter as shortcut in windows 

create bat file 
open in text editor 
and copy this 


```
call activate main
jupyter lab 	
```

for work this command there is need to add anaconda to windows variable path like above  link about Conda command is not recognized on Windows 10

also install jupyter lab for env 

be careful the main env it is not default env the default env name is base

for run jupyter notebook change the code tto this : 

```
call activate main
jupyter notebook 	
```

for change the default start up folder follow this link [How to change the Jupyter start-up folder](https://stackoverflow.com/questions/35254852/how-to-change-the-jupyter-start-up-folder)

or if you want set the path in script use this : 


```
call activate main
jupyter lab --notebook-dir E:\code 
```
