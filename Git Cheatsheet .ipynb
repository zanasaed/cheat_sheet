## Staging files
Stages a single file

`git add file1.js`  

Stages multiple files

`git add file1.js file2.js`  

Stages with a pattern

`git add *.js`

Stages the current directory and all its content

`git add .`  

## Viewing the status

**Full status**

`git status` 

**Short status**

`git status -s`


## Committing the staged files

Commits whit a one-line message 

`git commit -m "Message"`

Opens the default editor to type a long message 

`git commit`


##Skipping the stagging area

git commit -am "Message"

##Removing files 
instead of using Unix command to remove files and then git add remove file to delete files from work directory and stage directory we can use git command.

Removes from working directory and staging area 

`git rm file1.js`

Removes from staging area only

`git rm --cached file1.js` 

##Remove allready commited file and added to git ignore [linke](https://www.git-tower.com/learn/git/faq/ignore-tracked-files-in-git/)


```
# Remove the files from the index (not the actual files in the working copy)
$ git rm -r --cached .

# Add these removals to the Staging Area...
$ git add .

# ...and commit them!
$ git commit -m "Clean up ignored files"


```




## Renaming or moving files 
`git mv file1.js file1.txt` 

## Viewing the stage/unsataged changes 
Shows unstaged changes 
`git diff`

shows staged changes 
`git diff --staged` 

Same as the above 
`git diff --cached` 

##Viewing the history 

Full history 

`git log `

Summary 

`git log--oneline `

Lists the commits from the oldest to the newest 

`git log --reverse`

##viewing a commit 

Shows the gives commit 

`git show 921a2ff`

Shows the last commit

`git show HEAD`

Two steps before the last commit 

`git show HEAD~2`

Shows the version of file.js stored in the last commit 

`git show HEAD:file.js`

first copy commit id in list of commit (git log)
