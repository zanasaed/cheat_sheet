## **work with branch** 



**show all branch in project**


```
git branch
```


**made a brach**

```
git brach branch_name
```


**switche on branch**

```
git checkout branch_name
```

**merge branch with master**

first go on master then : 

```
git merge branch_name
```

**delete branch**

```
git branch -d branch_name
```

**for aplly change on origin (repository) :**

```
git push origin --delete branch_name
```

## **Log and show** 

**show history of commit**

```
git log
```


**if use commit id  in show command then can see details of commit** 


```
show commit_id
```



**show last commit**

```
git show HEAD
```



## **TAGS**

**we can use tag for tag a version to the code**

---

**for see history of tags :**
```
git tag
```

**add tag :**

```
git tag -a v# -m "this is the version 2.0 # is the number of version" 
```

**add tag to  spcific commit :**
first copy commit id in list of commit (git log)

```
git tag -a v1.2 commit_id -m " this is the version 1.2 made by zana"
```

**push all tags on orgin :**

```
git push orgin --tags or tag_name 
```
