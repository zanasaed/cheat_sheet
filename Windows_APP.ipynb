#### insert code insde word like IDE format


**insert code insde word like IDE format**

see this [link](https://stackoverflow.com/questions/387453/how-do-you-display-code-snippets-in-ms-word-preserving-format-and-syntax-highlig) and made macro base on this solution.

macro code it is like this:
```
Sub insert_code_with_format()
'
' insert_code_with_format Macro
'
'
    Selection.InlineShapes.AddOLEObject ClassType:="Word.OpenDocumentText.12", _
         FileName:="", LinkToFile:=False, DisplayAsIcon:=False
    Selection.Paste
    ActiveDocument.SaveAs2 FileName:= _
        "C:\Users\zana\AppData\Local\Microsoft\Windows\INetCache\Content.Word\~WRO1103.doc" _
        , FileFormat:=wdFormatXMLDocument, LockComments:=False, Password:="", _
        AddToRecentFiles:=True, WritePassword:="", ReadOnlyRecommended:=False, _
        EmbedTrueTypeFonts:=False, SaveNativePictureFormat:=False, SaveFormsData _
        :=False, SaveAsAOCELetter:=False, CompatibilityMode:=15
    ActiveWindow.Close
End Sub


```
